```python
eng = int(input("Enter marks for English: "))
maths = int(input("Enter marks for Maths: "))
sci = int(input("Enter marks for Science: "))
history = int(input("Enter marks for History: "))
geo = int(input("Enter marks for Geography: "))
hindi = int(input("Enter marks for Hindi: "))

total_marks = eng + maths + sci + history + geo + hindi
number_of_subjects = 6
average_marks = total_marks / number_of_subjects

print(f"\nTotal Marks: {total_marks}")
print(f"Average Marks: {average_marks:}")

if average_marks >= 90:
    grade = 'A+'
elif average_marks >= 80:
    grade = 'A'
elif average_marks >= 70:
    grade = 'B+'
elif average_marks >= 60:
    grade = 'B'
elif average_marks >= 50:
    grade = 'C'
else:
    grade = 'F'

print(f"Grade: {grade}")
```
