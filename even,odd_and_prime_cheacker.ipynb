```python
def is_even(number):
    """Checks if a number is even."""
    return number % 2 == 0

def is_odd(number):
    """Checks if a number is odd."""
    return number % 2 != 0

def is_prime(number):
    """Checks if a number is prime."""
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True
```

```python
try:
    user_input = input("Enter a number to check: ")
    num = int(user_input)

    even_status = "Even" if is_even(num) else "Not Even"
    odd_status = "Odd" if is_odd(num) else "Not Odd"
    prime_status = "Prime" if is_prime(num) else "Not Prime"

    print(f"Number {num}: {even_status}, {odd_status}, {prime_status}")
except ValueError:
    print("Invalid input. Please enter an integer.")
```
